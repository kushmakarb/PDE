The Fourier cosine series are given by 
$$
    f(x) = \frac{c_0}{2} \sum_{k=1}^{\infty} c_k \cos(k\pi x),
$$
where
$$
    c_k = 2\left<f(x),\cos(k\pi x) \right>, \ k=0,1,\ldots.
$$

a)

The value $c_0/2$ corresponds to the average value of $f(x)=1+x$ on $[0,1]$. So, we have

$$c_0 = 2 \int_0^1 (1+x) dx=3$$

We then have the following for $k = 1,2,\ldots$ 

$$c_k = 2 \int_0^1 (1+x) \cos(k\pi x) dx=2 \int_0^1 \cos(k\pi x) + x\cos(k\pi x) dx$$

$$=2 \int_0^1 \cos(k\pi x) dx + 2 \int_0^1 x \cos(k\pi x) dx  $$

$$=\frac{2}{k^2\pi^2} ( \cos{k\pi} - 1) $$

$$\left \{ \begin{array}
& -\frac{4}{k^2\pi^2} & \text{if} \quad k = 1,3, \ldots \\
0 & \text{if} \quad k = 2,4,\ldots
\end{array} \right.
$$

Therefore, the Fourier Cosine series is given by

$$f(x)= \frac{3}{2} - \sum_{k\in odd}^{\infty} \frac{4}{k^2\pi^2} \cos(k\pi x)$$


b) 

The value $c_0/2$ corresponds to the average value of $f(x)=x^2$ on $[0,1]$. So, we have

$$c_0 = 2 \int_0^1 x^2 dx=\frac{2}{3}$$

We then have the following for $k = 1,2,\ldots$ 

$$c_k = 2 \int_0^1 x^2 \cos(k\pi x) dx = \frac{2x^2}{k\pi}\sin{k\pi x}\big|_0^1 - \frac{4}{k\pi} \int_0^1 x\sin(k\pi x) dx $$

$$=\frac{4}{k^2\pi^2} \cos({k\pi})$$

$$= \frac{4(-1)^k}{k^2\pi^2} $$

Therefore, the Fourier cosine series is given by

$$f(x)=\frac{1}{3}+\sum_{n=1}^{\infty} \frac{4(-1)^k}{k^2\pi^2} \cos (k \pi x)$$

c)

The value $c_0/2$ corresponds to the average value of $f(x)=x(1-x)$ on $[0,1]$. So, we have

$$c_0 = 2 \int_0^1 x(1-x)dx=\frac{1}{3}$$

We then have the following for $k = 1,2,\ldots$ 

$$c_k = 2 \int_0^1 x(1-x) \cos(k\pi x) dx = 2 \int_0^1 x \cos(k\pi x) dx - 2 \int_0^1 x^2 \cos(k\pi x) dx$$

First we solve for part of the above 

$$2 \int_0^1 x \cos(k\pi x)dx = \frac{2x}{k\pi}\sin{k\pi x}\big|_0^1 - \frac{2}{k\pi} \int_0^1 \sin(k\pi x) dx $$

$$
\left \{ 
\begin{array}
& -\frac{4}{k^2\pi^2} & \text{if} \quad k = 1,3,5,\ldots \\
0 & \text{if} \quad k = 2,4,6,\ldots
\end{array} \right.
$$

Since we already know another part of the integral from part (b), so we now combine these two solutions to acquire our Fourier cosine series

$$c_k= 2 \int_0^1 x \cos(k\pi x) dx - 2 \int_0^1 x^2 \cos(k\pi x) dx$$


$$= \left \{ 
\begin{array}
& 0 & \text{if} \quad k = 1,3,5,\ldots \\
-\frac{4}{k^2\pi^2} & \text{if} \quad k = 2,4,6,\ldots
\end{array} \right.
$$

Therefore, the Fourier Cosine Series is given by 

$$f(x)=\frac{1}{6} - \sum_{k \in even}^{\infty} \frac{4}{k^2\pi^2} \cos(k\pi x)$$

```python
N = np.array([3, 10, 100])
x = np.linspace(0,1, 2*N.max())

for N_iter in N:
    C_N = 0*x
    S_N = 0*x
    for k in range(1,N_iter):
        C_N += 1.0/(2*k-1)**2*np.cos((2*k-1)*np.pi*x)
        S_N += ((-1.0)**k)/k*np.sin(k*np.pi*x)
    S_N *= -2.0/np.pi
    C_N *= -4.0/(np.pi**2)
    plt.figure(1)
    plt.plot(x,S_N, label='N='+str(N_iter))
    plt.figure(2)
    plt.plot(x,C_N + 0.5, label='N='+str(N_iter))

plt.legend(shadow=True) 
```

As outlined in Section 3.6 (in particular Eq. (3.52)), the formal solutions are given by

$$
    u(x,t) = \frac{c_0}{2} + \sum_{k=1}^\infty c_k e^{-(k\pi)^2t}\cos(k\pi x),
$$

where

$$
    c_k = 2\left< f(x), \cos(k\pi x)\right>, \ k=0,1,\ldots.
$$

a)

For the initial function $f(x)=\cos(14\pi x)$. We have

$$c_0 = 2\int_0^1 \cos(14\pi x) dx = 2 \frac {\sin(14 \pi x)}{14 \pi}\big|_0^1=0$$

We then have the following for $k = 1,2,\ldots$ 

$$c_k = 2\left< \cos(14\pi x), \cos(k\pi x)\right>$$

Thus $c_k = 0$ for $k\neq14$ and

$$c_{14} = \int_0^1 \cos^2(14\pi x)dx = \int_0^1 \cos^2(14\pi x)dx = \int_0^1 \frac{1+\cos(28\pi x)}{2} dx = \frac{1}{2}$$

Therefore, we have

$$u(x,t) = \frac{1}{2} e^{-(14\pi)^2 t} \cos(14\pi x)$$


b)

For the initial function $f(x)=\sin(\pi x)$. We have

$$c_0 = 2\int^1_0 \sin(\pi x)dx = \frac{4}{\pi}$$

We then have the following for $k = 1,2,\ldots$ 

$$c_k = 2\int^1_0 \sin(\pi x)\cos(k\pi x)dx = 2 \left(\frac{(-1)^k + 1}{\pi - \pi k^2}\right)$$

by doing integration by parts.

Therefore, we have

$$u(x,t) = \frac{4}{\pi} + 2 \sum_{k=2}^{\infty} \left(\frac{(-1)^k + 1}{\pi - \pi k^2}\right) e^{-(k\pi)^2 t}\cos(k \pi x)$$

c)

For the initial function $f(x)=x^3$. We have

$$c_0 = 2\int^1_0 x^3 dx = \frac{1}{2}$$

We then have the following for $k = 1,2,\ldots$ 

$$c_k = 2\int^1_0 x^3\cos(k\pi x)dx = \frac{(6(k\pi)^2 - 12)(-1)^k + 12}{(k\pi)^4}$$

by using integration by parts.

Therefore, we have

$$u(x,t) = \frac{1}{4} +  \sum_{k=1}^{\infty} \frac{(6(k\pi)^2 - 12)(-1)^k + 12}{(k\pi)^4} e^{-(k\pi)^2 t}\cos(k \pi x)$$

<h1><center> Exercise 3.9 Solution </center></h1>

A family of a particular solutions ${u_k}$ are found by separation of variables. Inserting the ansatz

$$u_k(x,t) = X_k(x)T_k(t)$$

into the above equation. This gives us

$$X_k(x)T_k^\prime(t) = X_k^{\prime\prime}(x)T_k(t)+X_k^\prime(x)T_k(t)$$

We then have

$$\frac{T_k^\prime(t)}{T_k(t)} = \frac{X_k^{\prime\prime}(x) + X_k^\prime(x)}{X_k(x)} = -\lambda_k$$

where $\lambda_k$ is a constant. We then have

$$X_k^{\prime\prime}(x) + X_k^\prime(x) + \lambda_kX_k(x) = 0$$

and 

$$T_k^\prime(t) + \lambda_k T_k(t) = 0$$

by doing simple algebraic manipulations.

We then have the chracteristic polynomial:

$$r^2 + r + \lambda_k = 0$$

We solve for r, then we have

$$r = \frac{1}{2} \pm \sqrt{\frac{1}{4} - \lambda_k}$$

$$
X_k(x) = ae^{\frac{-x}{2}}\cos((\sqrt{\lambda_k - \frac{1}{4}})x) + be^{\frac{-x}{2}}\sin((\sqrt{\lambda_k - \frac{1}{4}})x)
$$

Now, using the boundary conditions, we get

$$X_k(0) = a = 0$$

and 

$$X_k(1) = be^{ -1 / 2 } \text{sin}\left(\frac{\sqrt{\lambda_k - \frac{1}{4}}}{2}\right) = 0$$

Since we want nontrivial solution, we solve $\sin(\sqrt{\lambda_k - \frac{1}{4}}) = 0$. Then, we have 

$$\sqrt{\lambda_k - \frac{1}{4}} = k\pi \Rightarrow \lambda_k = k^2\pi^2 + \frac{1}{4} \quad \text{for} \quad k=1,2,\ldots$$

Hence, our solution is given by 

$$X_k(x) = b_ke^{\frac{1}{2}}\sin((\sqrt{\lambda_k - \frac{1}{4}})x) = b_ke^{\frac{1}{2}}\sin(k\pi x) $$

Now, we need to solve

$$T_k^\prime(t) + \lambda_k T_k(t) = 0$$

Therefore, our solution is given by

$$T_k(t) = be^{-\lambda_kt} = be^{(k^2\pi^2 + \frac{1}{4})t}$$

Now we let $c_k = ab_ke^\frac{1}{2}$, we then have our family of particular solution by

$$u_k(x,t) = c_k\sin(k\pi x)e^{(k^2\pi^2 + \frac{1}{4})t}$$
