# Exercise 2.4

Construct a solution of the following two-point BVP:
$$
    -u''(x) = f(x), \ x\in(0,1), \ u(0)=\alpha, \ u(1)=\beta,
$$
where $\alpha$ and $\beta$ are given real numbers.

# Solution:
    
We start by implementing the Fundamental Theorem of Calculus in which there is a constant $c_1$ such that

$$u(x)=c_1+\int_0^x u'(y)dy,$$

and similarly, we have a constant $c_2$ such that

$$u'(y)=c_2+\int_0^y u''(z)dz$$

This would be true for any twice differentiable function u. Now we assume that u satisfies our differential equation. Then, above expression implies that 

$$u'(y)=c_2-\int_0^y f(z)dz$$

Then we plug this into above u(x), and we get

$$u(x)=c_1+c_2-\int_0^x \big(\int_0^y f(z) dz \big)dy.$$

Now we define $F(y)=\int_0^y f(z) dz$ to rewrite the above expression in a convenient form.

We then get

$$\int_0^x \big (\int_0^y f(z) dz \big )dy=\int_0^x F(y)dy$$

$$=[yF(y)]_0^x -\int_0^x yF'(y)dy$$

$$=xF(x)-\int_0^x yf(y)dy$$

$$=\int_0^x (x-y)f(y)dy$$

Now we have

$$u(x)=c_1+c_2x-\int_0^x (x-y)f(y)dy$$

Note that $c_1$ and $c_2$ are arbitary constants and now we use boundary conditions $u(0)=\alpha, \ u(1)=\beta$

Then we get $\alpha=c_1$ and $c_2 = \beta - \alpha + \int_0^1 \! (1-y)f(y)dy$

Therefore, our solution is

$$u(x) = \alpha + [\beta - \alpha ]x + \big (x \int_0^1 \! (1-y)f(y)dy \big)  - \int_0^x  (x-y) f(y) dy$$









 

# Exercise 2.5

Find a Green's function for the following two-point BVP:
$$
    -u''(x) = (x), \ x\in(0,1), \ u(0)=0, \ u'(1) = 0.
$$
Is there a unique solution of this problem?

# Solution:
    
Using the general solution from the equation above we have

$$u(x) = c_2 + c_1 x - \int_0^x \bigg( \int_0^y  z dz \bigg) dy$$

$$= c_2 + c_1 x - \int_0^x \frac{y^2}{2} dy$$ 

$$= c_2 +c_1 x -\frac{x^3}{6}$$

Now applying the boundary condition we get
$$u(x) = \frac{x}{6} - \frac{x^3}{6}$$

From exercise 2.3 we have the following 

$$ G(x,y) = \left\{ \begin{array}{rr}
                                y ,&  y\leq x.\\
                                x ,& x<y , 
                            \end{array}
                    \right.$$
this satisfies all the properties of a green function.

We now look further into our function

$$\int_0^1  G(x,y)f(y) dy = \int_0^x  y(1-x)y dy + \int_x^1 x(1-y)y dy $$

Now solving these integral on right get us to

$$\int_0^1  G(x,y)f(y) dy = \frac{x}{6} - \frac{x^3}{6}$$

So, we have $u(x)=\int_0^1  G(x,y)f(y) dy$.

Now we look into to see if the solution is unique by doing the following

Let $v(x)$ and $z(x)$ be both solutions to boundary value problem above. Then, we have $v(x)-z(x)$
satisfying the above boundary value problem with $f(x)=0$ for all $x\in(0,1)$ Thus, we have

$$v(x)-z(x)=\int_0^1 G(x,y)* 0\ dy=0$$

Hence, $v(x)=z(x)$. Therefore, our solution $u(x)$ must be unique.

# Exercise 2.9

Show that the BVP problem
$$
    -u''(x)=f(x), \ x\in(0,1), \ u(0)=u(1)=0
$$
is stable with respect to perturbations in the data.

# Solution:
    
Since $f$ is continuous on $[0,1]$. Then we have the following from 2.4

$$u(x) = x \int_0^1 (1-y)f(y) dy - \int_0^x  (x-y) f(y) dy$$

We then consider the equation

$$- z''(x) = f(x + \theta),  x\in(0,1),   z(0) = z(1) = 0$$

where $\theta$ is a small perturbation in the data. We then have following as our solution

$$z(x) = x \int_0^1 (1-y)f(y+\theta) dy - \int_0^x (x-y) f(y+\theta) dy$$

Now in the following steps we try to show the stability of the solution by using reverse triangular inequality and our regular similar $\theta-\epsilon$ approach. We are basically going to do a simple $\theta-\epsilon$ proof to show that the solution of the above differential equation is stable when we have a small perturbation $\theta$ in the data.

$$\big| z(x) - u(x) \big| = \bigg| x \int_0^1  (1-y)f(y+\theta) dy - \int_0^x (x-y) f(y+\theta) dy - x \int_0^1 (1-y)f(y) dy + \int_0^x (x-y) f(y) dy \bigg|$$

$$= \bigg| x \int_0^1 (1-y)\big( f(y+\theta) -f(y)\big) dy - \int_0^x  (x-y)\big(f(y+\theta) - f(y) \big) dy \bigg|$$

Now by the reverse triangular inequality, we have
$$\big| z(x) - u(x) \big| \leq   x \int_0^1 \!\Big| (1-y)\big( f(y+\theta) -f(y)\big)\Big| dy + \int_0^x \Big| (x-y)\big(f(y+\theta) - f(y) \big) \Big| dy$$

$\forall$$y \in (0,1)$ then we can have $|1-y| \leq 1 $ and for any $y \in(0,x)$ for $x \in (0,1)$ we can see that
$|x-y| \leq |x| = x$

Therefore, it follows that we have

$$\big| z(x) - u(x) \big| \leq x \int_0^1 \!\Big| f(y+\theta) -f(y) \Big| dy + \int_0^x \! \Big| x\big(f(y+\theta) - f(y) \big) \Big| dy$$

Then, $\exists$ $v \in (0,1)$ and $v_x \in (0,x)$ such that

$$\int_0^1 \Big| f(y+\theta) -f(y) \Big| dy \leq \big|f(v+\theta) - f(v)\big|$$

and

$$\int_0^x \Big| x\big(f(y+\theta) - f(y) \big) \Big| dy \leq \big|x\big(f(v_x +\theta) - f(v_x)\big)\big|x = x^2 \big|f(v_x+\theta)-f(v_x)\big|$$

So, $f$ is continuous in the compact set, thus it is uniformly continuous by a theorem (referenced undergraduate analysis book) because continousness in the compact set directly imples uniform continuity. So, we know for $\epsilon>0$ if we make $\theta$ small enough, then for any $|x-y| < \theta$ we then have

$$\big|f(x) - f(y)\big| < \epsilon$$

Therefore, we have 

$$\big|z(x) - u(x)\big| \leq  x\big|f(v+\theta) - f(v)\big| + x^2 \big|f(v_x+\theta)-f(v_x)\big| <  x\epsilon + x^2\epsilon \leq  \epsilon + \epsilon  =  2\epsilon$$

Now we can say that the even the small peturbation in the data does not change the solution by a whole alot or in other words there is also a small change in solution when the data is perturbated in small rate. This means that our solution is stable with respect to the perturbations in the data.

a) Explain the derivation of (2.55) given by
$$
    \int_a^b F(x)\, dx \approx h\left(\frac{F(a)}{2} + \sum_{i=1}^n F(x_i) + \frac{F(b)}{2}\right).
$$

# Solution:
    
The trapizoidal is a way of approximating definite integrals. In this, the region under the graph is approximated by calculating the areas of the trapizoids. Suppose $F(x)$ is a continuous function with $x \in [a,b]$. We let n be any arbitary natural number and we consider evenly spaced grid of $[a,b]$ 

$$a = x_0 < x_1 < x_2 < \ldots < x_n < x_{n+1} = b$$

We then let $h=\frac{(bâˆ’a)}{n+1}$ then we have $x_k = a+kh$ for each $k\in(0,1,\ldots,n+1)$. We then can approximate our integral with n+1 trapizoids. We then have for any $i \in (1,2,\ldots,n+1)$ the area $A_i$ of the $i^\text{th}$ trapezoid can be calculated by

$$A_i = hF(x_{i-1}) + h\frac{1}{2}\big(F(x_i) - F(x_{i-1}\big) = h\frac{1}{2}\big(F(x_i) + F(x_{i-1})\big)$$

We can use above information to approximate our integral

$$\int_a^b F(x)dx \approx \sum_{i=1}^{n+1} A_i = h \sum_{i=1}^{n+1}\frac{1}{2}\big(F(x_i) + F(x_{i-1}) \big)$$

Then, we have

$$= h\Bigg( \frac{F(x_1)}{2} + \frac{F(x_0)}{2} + \frac{1}{2}\sum_{i=2}^n F(x_i) + \frac{1}{2}\sum_{i=2}^n F(x_{i-1}) + \frac{F(x_{n+1})}{2} + \frac{F(x_n)}{2}\Bigg)$$

Which it then lead us to

$$= h\Bigg( \frac{F(a)}{2} + \frac{F(x_1)}{2} + \frac{1}{2}\sum_{i=2}^n F(x_i) + \frac{1}{2}\sum_{i=1}^{n-1} F(x_i) + \frac{F(x_n)}{2} + \frac{F(b)}{2}\Bigg) $$

It follows that 

$$= h\Bigg( \frac{F(a)}{2} + \frac{1}{2}\sum_{i=1}^n F(x_i) + \frac{1}{2}\sum_{i=1}^n F(x_i) + \frac{F(b)}{2}\Bigg)$$

Thus, we have

$$= h\Bigg( \frac{F(a)}{2} + \sum_{i=1}^n F(x_i) + \frac{F(b)}{2}\Bigg) $$

as desired.


b) Write a procedure that, for a given $a, b, F(x),$ and $n,$ computes the approximation defined by (2.55:part a).

```python
    
def F(j):
    F = j**5
    return F

def G(j):
    G = np.sqrt(np.absolute(j - 0.5))
    return G

def TrapInt(a , b , F, n):
    x = np.linspace(a , b , n+2)
    f = np.zeros(n+1)
    g = 0
    h = (b-a)/(n+1)
    
    for i in range(1,n+1):
        f[i] = F(x[i])
        g += h*f[i]
    
    g = g + (h*F(a) / 2) + (h*F(b) / 2)
    
    return g
```

```python
def approxu(s , t , f1 , n):
    
    u = np.zeros(n+2)
    
    h = (t - s)/(n+1)
    
    y = np.linspace(s , t , 2*n +3)
    
    x = np.linspace(s , t , n +2)

    f = np.zeros(2*n +3)
    
    c = np.zeros(n+2)
    
    d = np.zeros(n+2)
    
    for i in range(0 , 2*n +3):
        
        f[i] = f1(y[i])
        
    for i in range(0 , n+1):
        
        c[i+1] = c[i] + (h/2)*(f[2*i] + 2*f[2*i +1] + f[2*i +2])
        
        d[i+1] = d[i] + (h/2)*(x[i]*f[2*i] + 2*y[2*i +1]*f[2*i +1] + x[i+1]*f[2*i +2])
      
    for i in range(1 , n+1):
        
        u[i] = x[i]*(c[n+1] - d[n+1]) + d[i] - x[i]*c[i]
        
    return u
```
